```python
# for matching signals with job_titles
title=[]
for profile in profiles:
    time.sleep(random.randint(5, 10))
    driver.get(profile)
    job_title=driver.find_elements_by_xpath("//h2[@class='mt1 t-18 t-black t-normal break-words']")
    title_single=[]
    for i in job_title:
        title_single.append(i.get_attribute("innerHTML"))
    title.append(title_single)
    
```

```python
title
```

```python
title_list = [ item for elem in title for item in elem]
```

```python
title_list
```

```python
import re
```

```python
# for matching signals with experience
exp=[]
for profile in profiles:
    time.sleep(random.randint(5, 10))
    driver.get(profile)
    experience=driver.find_elements_by_tag_name("p")
    exp_single=[]
    for i in experience:
         cleanr = re.compile('<.*?>|&([a-z0-9]+|#[0-9]{1,6}|#x[0-9a-f]{1,6});,.')
         cleantext = re.sub(cleanr, '',i.get_attribute("innerHTML")).strip()
         exp_single.append(cleantext)
    exp.append(' '.join(exp_single))
        
        
        
        
```

```python
exp
```

```python
pro_list=url[3] + " "+ url[4] + " " + url[5]
```

```python
pro_list
```

```python
pro_signals=pro_list.values.tolist()
```

```python
#Cleaning the tags, category and needs data(signals) 
edit_list=[]
for i in range(len(pro_signals)):
    edit = pro_signals[i].replace('('," ").replace(')'," ")
    edit_list.append(edit)
```

```python
edit_list
```

# Relevant signals matching

```python
# Matching signals with job_title and experience
signals_dictionary={}    
for i in range(len(exp)):
    exp_words = exp[i].lower().split()
    title_words = title_list[i].lower().split()
    signals_words = edit_list[i].lower().split()
    common_exp = set(exp_words).intersection( set(signals_words) )
    common_title = set(title_words).intersection( set(signals_words) )
    def find_distinct(common):
        distinct=list()
        for element in common:
            if element not in ["in","of","and","the","or"]:
                distinct.append(element)
        return(distinct)        
    signals_dictionary[profiles[i]]={"relevant signals from experience":find_distinct(common_exp),
                                      "relevant signals from title":find_distinct(common_title)
                                     } 

    
                   
```

```python
signals_dictionary
```

```python
# for experience data
exp_dictionary={}
for i in range(len(profiles)):
    exp_dictionary[profiles[i]]={"work_exp":exp[i]
        
    }
   
```
